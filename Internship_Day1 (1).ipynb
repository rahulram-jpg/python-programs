```python
def max_of_three(a, b, c):
  if a >= b and a >= c:
    return a
  elif b >= a and b >= c:
    return b
  else:
    return c
max_of_three(1, 2, 3)
```

```python
def vowels(s):
  a=0
  v=["a","e","i","o","u"]
  for i in s:
    if i in v:
      a+=1
  return a
str=input("enter a string=")
S=str.lower()
print("vowels=",vowels(S))
```

Q2

```python
def prime(n):
  if n==2:
    return True
  for i in range(2,(int(n**0.5)+1)):
    if n%i==0:
      return False
    return True
num=int(input("enter a number="))
print(prime(num))
```

Q4

```python
def oddcount(n):
    count = 0
    for i in str(n):
        if int(i) % 2 != 0:
            count += 1
    return count

n = int(input("Enter a number: "))
print(oddcount(n))
```

Q5

```python
def remove_spaces(s):
    result = ""
    for ch in s:
        if ch != " ":
            result += ch
    return result

text = input("Enter a string: ")
print(remove_spaces(text))
```

```python
def smallest(a, b, c):
    if a <= b and a <= c:
        return a
    elif b <= a and b <= c:
        return b
    else:
        return c

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))

print("Smallest number =", smallest(x, y, z))
```
