```python
a = np.zeros([2,2])
b = np.ones([1,4])
c = np.ones([2,2])
d = np.eye(2)
e = np.eye(4)
f = np.ones([3,3])
g = np.eye(3)
h = np.ones([3,1])

print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
```

```python
a1=np.array([1,2,3,])
a2=np.array([3,4,5])
print(a1+a2)
print(a1-a2)
print(a1/a2)
print(a1*a2)
```

```python
arr=np.array([[1,2,3], [4,5,6]])

mean=np.mean(arr)
print(mean)
sum_rows=np.sum(arr, axis=0)
print(sum_rows)
max_cols=np.max(arr, axis=0)
print(max_cols)
varience=np.var(arr)
print(varience)
```
