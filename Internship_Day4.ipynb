```python
numerical_features=df.select_dtypes(include=['float64','int64'])
skewness=numerical_features.skew().sort_values(ascending=False)
print("Skewness of Numerical Features:")
print(skewness)
```

```python
# Plot distributions of numerical features
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_features.columns, 1):
    plt.subplot((len(numerical_features.columns) // 3) + 1, 3, i)
    sns.histplot(df[column].dropna(), kde=True, bins=30, color='blue')
    plt.title(f'{column} (Skewness: {skewness[column]:.2f})', fontsize=10)
    plt.xlabel(column, fontsize=8)
    plt.ylabel('Frequency', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

encoded_data = df.copy()

# 1. Label Encoding for 'Education'
label_encoder= LabelEncoder()
encoded_data['Education_Label'] = label_encoder.fit_transform(encoded_data['Education'])

# 2. One-Hot Encoding for 'Department'
one_hot_encoded = pd.get_dummies(encoded_data['Department'], prefix='Department')
encoded_data = pd.concat([encoded_data, one_hot_encoded], axis=1)

# Since 'Performance_Rating' is ordinal, we can create a mapped column for demonstration.
# 3. Ordinal Encoding For Performance_Rating
pclass_mapping = {1: 'First', 2: 'Second', 3: 'Third', 4: 'Fourth', 5: 'Fifth'}
encoded_data['Performance_Rating_Category'] = encoded_data['Performance_Rating'].map(pclass_mapping)

# Display the resulting dataset with relevant encoded columns
print(encoded_data[['Education', 'Education_Label', 'Department', 'Department_Finance', 'Department_HR', 'Department_IT', 'Department_Marketing', 'Performance_Rating', 'Performance_Rating_Category']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
#scikitlearn
# Load the dataset
file_path = '/content/preprocessing_practice_dataset.csv'
preprocessing_practice_dataset = pd.read_csv(file_path)

# Select numerical columns for scaling
numerical_columns = ['Employee_ID', 'Performance_Rating']

# Drop rows with missing values for simplicity
preprocessing_practice_dataset.dropna(subset=numerical_columns, inplace=True)

# Initialize scalers
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

# Apply Min-Max Scaling
preprocessing_practice_dataset['Employee_ID_MinMax'] = minmax_scaler.fit_transform(preprocessing_practice_dataset[['Employee_ID']])
preprocessing_practice_dataset['Performance_Rating_MinMax'] = minmax_scaler.fit_transform(preprocessing_practice_dataset[['Performance_Rating']])

# Apply Standard Scaling
preprocessing_practice_dataset['Employee_ID_Standard'] = standard_scaler.fit_transform(preprocessing_practice_dataset[['Employee_ID']])
preprocessing_practice_dataset['Performance_Rating_Standard'] = standard_scaler.fit_transform(preprocessing_practice_dataset[['Performance_Rating']])

# Display scaled features
preprocessing_practice_dataset[['Employee_ID', 'Employee_ID_MinMax', 'Employee_ID_Standard']].head()
```
